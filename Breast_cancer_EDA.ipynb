```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv('/content/breast-cancer.csv')
```

```python
## To check the datatype of each entries in data set
dataset.info()
```

```python
## id column is not important for analysis, hence it's removed from dataset.
dataset.drop(columns="id", axis=1,inplace=True)
```

# ## Descriptive summary of Dataset :
1. Check Mean, median(50% statistical value) are close to each other or not
2.  If Mean value is higher than median then the graph will show right skewed pattern
3. On the otherhand, Median value is higher than mean value, then the graph will show left skewed pattern

```python
# malignant and begnin types cancer
dataset['diagnosis'].value_counts()
```

## Applying Numerical Value to Object in Diagnosis column
1. Encoding taget column
2. Giving values to diagnosis column 1 or 0 to Benign or Malignant
3. Transfer all objects to labels with value
4. Create a target new column
5. remove diagnosis column

```python
label_encode=LabelEncoder()
labels=label_encode.fit_transform(dataset ['diagnosis'])
dataset['target']=labels
dataset.drop(columns='diagnosis',axis=1, inplace=True)
```

```python
# created target column, removed diagnosis object
dataset.head(10)
```

```python
## Benign type =0, Malignant=1, category
dataset.target.value_counts()
```

```python
dataset.groupby('target').mean()
```

# Inferene: What we learned from this EDA?!!
We observed clearly that the mean value of Malignant tumor cases are higher for all features than benign tumor type

# SUMMARY
1. No missing values found in data set
2. All values were continuous numerical type except the target column
3. Mean value was slightly higher than Median value (50%), hence graph may show rightly skewed  pattern.
4. Slightly imbalance dataset, begnin type (0) cases were more numbers than malignant cases.
5. Mean value for malignant cases were higher than the benign case observed from Groupby function.
